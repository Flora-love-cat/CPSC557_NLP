# noisy channel model

## definition

Noisy Channel Model

- come from information theory and communication systems. 

- transmitting information through a communication channel (encoder/transmitor) that introduces noise or errors. 

- goal: reconstruct input $X$ given the correputed output $Y$ by MLE or Bayesian inference.

    $$
    \hat X = \arg\max_{X} P(X|Y) = \arg\max_{X} P(Y|X) P(X)
    $$

    $P(Y|X)$: likelihood of output message given the input message
    
    $P(X)$: prior probability of the input message 


## application

Applications: Handwriting recognition, speech recognition, machine translation, Spelling correction, Text summarization, Text generation.

Example: machine translation

- translation model $P(Y|X)$ represents likelihood of generating target language Y given source language X

- language model $P(X)$ represents prior probability of input sentence in source language.



Example: speech recognition

- Input: Written English (X) - This is the original message that you want to communicate.

- Encoder: garbles the input from written English (X) to spoken English (Y).

- intermediate Output: Spoken English (Y) - This is the received message, which might be different from the original input message due to the noise introduced by the channel.

- Decoder: recover the Written English (X) from Spoken English (Y) 

- output: Written English (X') - reconstructed/denoised original message.
